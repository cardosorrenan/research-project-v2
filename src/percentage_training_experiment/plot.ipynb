```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(20, 6), dpi=80)

# Read the CSV file
data = pd.read_csv("./results_exp3.csv", delimiter=";")

# Extract the two columns
x = data["quantity_segments"]
acc = data["acc"]
acc = acc.apply(lambda x: x if isinstance(x, float) else float(x.split("+-")[0]))
spe = data["spe"]
spe = spe.apply(lambda x: x if isinstance(x, float) else float(x.split("+-")[0]))
sen = data["sen"]
sen = sen.apply(lambda x: x if isinstance(x, float) else float(x.split("+-")[0]))


# Create line plot using Seaborn
sns.lineplot(x=x, y=acc)
sns.lineplot(x=x, y=spe)
sns.lineplot(x=x, y=sen)

# Add ticks and vertical lines
for i in range(len(x)):
    plt.axvline(x=x[i], color="gray", alpha=0.1)
    plt.text(
        x=x[i], y=min(acc), s="|", color="gray", alpha=0.5, va="bottom", ha="center"
    )


plt.legend(labels=["Acurácia", "Especificidade", "Sensibilidade"])
# Display the plot

plt.show()
```
