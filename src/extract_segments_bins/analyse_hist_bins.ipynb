```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator
```

```python
base = np.load(f"bins/bins_50.npy")

afib = base[base[:, -1] == 1]

snr = base[base[:, -1] == -1]

np.random.seed(42)

data = np.mean(afib, axis=0)[0:-1]

sns.set(style="darkgrid")

fig, ax = plt.subplots(figsize=(12, 6))
bar_color = (0.2, 0.4, 0.6)

sns.barplot(x=np.arange(0, 2500, 50), color=bar_color, y=data, ax=ax)

ax.xaxis.set_major_locator(MultipleLocator(5))
ax.set_yticklabels(ax.get_yticks(), fontdict={"fontsize": 13})
ax.set_xticklabels(ax.get_xticks().astype(int) * 50, fontdict={"fontsize": 13})

ax.set_ylabel("Média dos histogramas (FA)", fontdict={"fontsize": 15})
ax.set_xlabel("ms", fontdict={"fontsize": 15})

plt.show()
```
